```python
from google.colab import files
import numpy as np

print("Upload your 15-column data file")
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Load data (try with different delimiters)
try:
    # Attempt comma delimiter first
    data = np.genfromtxt(filename, delimiter=',', encoding='utf-8-sig')
    if data.ndim == 1 or data.shape[1] < 3:
        # If failed, try space/tab delimiter
        data = np.genfromtxt(filename, delimiter=None, encoding='utf-8-sig')
except:
    raise ValueError("Could not load file as 15-column data. Check delimiter and column count.")

# Verify we have enough columns
if data.ndim == 1 or data.shape[1] < 3:
    raise ValueError(f"File must have 15 columns, but found {data.shape[1] if data.ndim > 1 else '1'} columns")

# Your original variable assignments
x = data[:,0]   # Column 1 # PHASE
y = data[:,1]   # Column 2 # MAG
x1 = data[:,2]  # Column 3
y1 = data[:,3]  # Column 4
x2 = data[:,4]  # Column 5
y2 = data[:,5]  # Column 6
x3 = data[:,6]  # Column 7
y3 = data[:,7]  # Column 8
x4 = data[:,8]  # Column 9
y4 = data[:,9]  # Column 10
# error bars:
flux_err = data[:,10] # ERROR
flux_err1 = data[:,11]
flux_err2 = data[:,12]
flux_err3 = data[:,13]
flux_err4 = data[:,14]
```

```python
import matplotlib.pyplot as plt

# Create plot using plt (instead of fig,ax)
plt.figure(figsize=(8, 4))  # Set figure size (optional)

# Plot all lines
plt.plot(x, y, 'o', linewidth=2, color='seagreen', label='2/17/2025')
plt.plot(x1, y1, 'o',linewidth=3,color='royalblue', label='2/27/2025')
plt.plot(x2, y2, 'o',linewidth=3,color='tomato', label='3/17/2025')
plt.plot(x3, y3, 'o',linewidth=3,color='blueviolet', label='4/14/2025')
plt.plot(x4, y4, 'o',linewidth=3,color='hotpink', label='4/16/2025')

# Add error bars
#plt.errorbar(x, y, yerr=flux_err, fmt='.', color='black')
plt.errorbar(x, y, yerr=flux_err,fmt='none', ecolor='black',elinewidth=1,capsize=2)
plt.errorbar(x1, y1, yerr=flux_err1,fmt='none', ecolor='black',elinewidth=1,capsize=2)
plt.errorbar(x2, y2, yerr=flux_err2,fmt='none', ecolor='black',elinewidth=1,capsize=2)
plt.errorbar(x3, y3, yerr=flux_err3,fmt='none', ecolor='black',elinewidth=1,capsize=2)
plt.errorbar(x4, y4, yerr=flux_err4,fmt='none', ecolor='black',elinewidth=1,capsize=2)

# Formatting
plt.title('T = 0.6470 Days, V-Band',fontsize=14, fontweight='bold')
plt.xlabel('Phase')
plt.ylabel('Magnitude')
#legend = plt.legend()
plt.ylim(8.89,8.81)
plt.xlim(0, 1)

# Save figure
plt.savefig('V-0.647.png',
           dpi=300,
           bbox_inches='tight',
           facecolor='white')

# Show plot
plt.show()
```

```python
# Save JUST the legend as PNG
fig_legend = plt.figure(figsize=(3, 1))  # Adjust size as needed
plt.figlegend(*plt.gca().get_legend_handles_labels(),
              loc='center',
              frameon=False)  # Recreate legend in a new figure
plt.axis('off')  # Hide axes
fig_legend.savefig('legend.png',
                  dpi=300,
                  bbox_inches='tight',
                  transparent=True)  # Remove background
plt.close(fig_legend)
```
